NLTK = Natural Language Toolkit — это огромная библиотека на Python для работы с текстом и языком.
Она популярна в академической среде, часто используется в курсах по NLP.

В ней есть:
токенизация (разбить текст на слова/предложения),
POS-тэггинг (определение частей речи),
парсинг предложений,
частотный анализ,
готовые словари и корпуса текстов (например, "Гамлет" или газетные статьи),
куча обучающих датасетов для экспериментов.

```python
from g2p_en import G2p
import nltk
# Скачиваем недостающие ресурсы (если нужно.  1 раз)  ^


# ?? для проставки английских УДАРЕНИЙ нужен контекст (перевести русские слова в порошке на английский)
#  + для проставки русских УДАРЕНИЙ нужен контекст (перевести английские слова в порошке на русский)
""" g2p_en берёт ударения из CMU Pronouncing Dictionary:
1 → главное ударение
2 → побочное ударение (слабее, но заметное)  !! 
0 → безударная
"""

# Скачиваем недостающие ресурсы (если нужно.  1 раз)  v
# nltk.download("punkt")
# nltk.download("averaged_perceptron_tagger_eng")
# nltk.download("punkt_tab")  токенизатор
 
# Иногда нужны ещё:      
""" 2025
# nltk.download("cmudict")   # словарь произношений (часто нужен для фонем)
# nltk.download("universal_tagset")  # универсальная разметка частей речи
# nltk.download("wordnet")  # лемматизация (часто нужна для фонем)
# nltk.download("omw-1.4")  # лемматизация (часто нужна для фонем)
# nltk.download("maxent_ne_chunker")  # для именованных сущностей
# nltk.download("words")  # для именованных сущностей
# nltk.download("tagsets")  # список всех тегов частей речи"""

g2p = G2p()

# ??  найти аналогично  фонемы для немецкого и французского языка  и латыни (с ударениями тоже)
# +  пакет для перевода чисел в слова ... в контексте ??
# ??   и РУССКОГО ?
#   КАРТА ФОНЕМ   !!   CMUdict, которые использует g2p_en
ARPABET2CYR = {
    # Гласные 19
    "AA": "а", "AE": "э", "AH": "а", "AO": "о",
    "AW": "ау", "AY": "ай", "EH": "е", "ER": "эр",
    "EY": "ей", "IH": "и", "IY": "и", "OW": "оу",
    "OY": "ой", "UH": "у", "UW": "у",
    "AX": "а", "IX": "и", "UX": "у", "AXR": "ар",

    # Согласные 24
    "B": "б", "CH": "ч", "D": "д", "DH": "з",
    "F": "ф", "G": "г", "HH": "х", "JH": "дж",
    "K": "к", "L": "л", "M": "м", "N": "н",
    "NG": "нг", "P": "п", "R": "р", "S": "с",
    "SH": "ш", "T": "т", "TH": "с", "V": "в",
    "W": "у", "Y": "й", "Z": "з", "ZH": "ж"
}

VOWELS = {  # 18 ?
    "AA", "AE", "AH", "AO", "AW", "AY", "EH", "ER",
    "EY", "IH", "IY", "OW", "OY", "UH", "UW", "AX", "IX", "UX"
}

COMMON_W_V = {            # дополнить список ???
    "west", "win", "wine", "wood", "wolf",
    "woman", "world", "war", "was", "want", "white"
}


def strip_stress(ph):
    """Убирает ударение: AH1 → AH, ER0 → ER."""
    return ''.join([c for c in ph if not c.isdigit()])


def smart_w(word, phonemes, i):
    next_ph = phonemes[i+1] if i+1 < len(phonemes) else None
    if word.lower() in COMMON_W_V:
        return "в"
    if next_ph in VOWELS:
        return "у"
    return "у"


def phonemes_to_cyr(word):
    phonemes = [strip_stress(ph) for ph in g2p(word) if strip_stress(ph).isalpha()]
    result = []
    for i, ph in enumerate(phonemes):
        if ph == "W":
            result.append(smart_w(word, phonemes, i))
        else:
            result.append(ARPABET2CYR.get(ph, ph))
    return "".join(result)


# тесты
print(g2p("1"))  # ['M', 'AH1', 'DH', 'ER0']

print(phonemes_to_cyr("this"))     # зис
print(phonemes_to_cyr("mother"))   # мазер
print(phonemes_to_cyr("with"))     # уис
print(phonemes_to_cyr("chemistry"))        #   
print(phonemes_to_cyr("west"))     # вест
print(phonemes_to_cyr("ex"))               #   
print(phonemes_to_cyr("water"))    # уотер
print(phonemes_to_cyr("twist"))    # туист
```

```python
import subprocess


def get_espeak_voices(exe_path=r"C:\Program Files\eSpeak NG\espeak-ng.exe"):
    """Возвращает список доступных голосов espeak-ng как (код, название)."""
    result = subprocess.run(
        [exe_path, "--voices"],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        encoding="utf-8"
    )

    voices = []
    for line in result.stdout.splitlines():
        parts = line.split()
        if len(parts) < 4 or parts[0] == "Pty":
            continue
        code = parts[1]         # язык (например fr-fr, de, ru, la)
        name = parts[3]         # имя голоса (например French_(France))
        voices.append((code, name))

    return voices


if __name__ == "__main__":
    voices = get_espeak_voices()
    print(f"{'Код':<10} | Название")
    print("-" * 40)
    for code, name in voices:
        print(f"{code:<10} | {name}")
```

```python
# ----------------------------
# сырая фунция на будущее  (для Адаптированной орфографии) для  .py
# сомниваюсь что она нужна
# ----------------------------

def ipa_to_cyrillic(ipa: str, adapt: bool = False) -> str:
    base = ipa_to_cyr(ipa)  # "спайде"        см _translit.py
    if adapt:
        return russian_adaptation(base, ipa)  # "спайдер"
    return base
def ipa_to_cyr(ipa: str) -> str:    # заглушка
    return ipa


# 1 вариант от местного ии


def russian_adaptation(translit: str, ipa: str) -> str:
    """
    Применяет правила адаптации к русской орфографии.
    translit: строка в приближённой кириллице (из ipa_to_cyr)
    ipa: оригинальная IPA строка (для контекста)
    """
    # Пример простых правил адаптации:
    # 1. Если слово заканчивается на согласный + "е" без ударения, заменить "е" на "ер"
    if translit.endswith("е") and not "+е" in translit:
        if len(translit) >= 2 and translit[-2] not in "аеёиоуыэюя":
            translit = translit[:-1] + "ер"
    return translit

# 2 вариант от GPT


def russian_adaptation(base: str, ipa: str) -> str:
    res = base

    # --- правило: конечный "а" от schwa в позиции r → "ер"
    # spider spˈaɪdə → спайда → спайдер
    # printer prɪntər → принта → принтер
    if ipa.endswith(("ə", "ər", "əɹ")) and res.endswith("а"):
        res = res[:-1] + "ер"

    # --- точечные фиксы (накапливать словарь)
    # добавлять по мере необходимости или находить закономерности
    fixes = {
        "мэниджа": "менеджер",
        "фэмилиа": "фамилия",
    }

    for k, v in fixes.items():
        if res == k:
            res = v

    return res
```

```python
import re

SUPPORTED = {"en", "de", "fr", "es", "it", "la"}


def detect_word_lang(word):
    w = word.lower()

    # Эвристики
    if re.search(r"[äöüß]", w):
        return "de"
    if re.search(r"[éèêëàâîïç]", w):
        return "fr"
    if "ñ" in w:
        return "es"

    # fallback
    return "en"


def detect_languages(text):
    words = re.findall(r"\w+", text)
    return [(w, detect_word_lang(w)) for w in words]


# пример
print(detect_languages("Hola my friend guten Tag"))
```

```python
import subprocess
from typing import Optional

# g2p_en
try:
    from g2p_en import G2p
    g2p = G2p()
except ImportError:
    g2p = None


def espeak_ipa(word: str, lang: str = "en") -> str:
    """Вызывает espeak-ng для получения IPA"""
    try:
        result = subprocess.run(
            ["espeak-ng", "-q", "--ipa", "-v", lang, word],
            capture_output=True, text=True, encoding="utf-8"
        )
        return result.stdout.strip()
    except Exception as e:
        return ""


def g2p_en_ipa(text: str) -> str:
    """G2P для английского"""
    if not g2p:
        return ""
    try:
        out = g2p(text)
        # g2p возвращает список токенов с пробелами
        return "".join(out).replace(" ", "")
    except Exception:
        return ""


def word_to_ipa(word: str, lang: str, use_g2p: bool = True) -> Optional[str]:
    """
    Универсальная функция:
    - для английского: g2p_en (с fallback на espeak)
    - для остальных: espeak-ng
    """
    if lang == "en" and use_g2p and g2p:
        ipa = g2p_en_ipa(word)
        if ipa:  # если получилось
            return ipa

    # fallback на espeak
    ipa = espeak_ipa(word, lang)
    return ipa if ipa else None


# ---------------------------
# Тест
# ---------------------------
tests = [
    ("price", "en"),
    ("chemistry", "en"),
    ("Mutter", "de"),
    ("fête", "fr"),
    ("hola", "es"),
    ("amico", "it"),
    ("mater", "la")
]

for word, lang in tests:
    ipa = word_to_ipa(word, lang, use_g2p=True)
    print(f"{lang:2} {word:10} -> {ipa}")
```
