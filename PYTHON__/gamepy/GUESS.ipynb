```python
#  чужое

def playing():

    # импортируем некоторые модули
    import random
    import time

    # Величины для отслеживания рекордов и кол-ва игр
    mx = -1
    mn = 99999999999
    counter_play = 0

    # Небольшая функция для оформления статистики
    def decoration():

        print("-" * 20)

    # Название игры
    print("Числовая угадайка!")

    # Указываем свое имя
    name = input("Введите свое имя: ")

    # Приветствие и объяснение, как перейти на определенный режим
    print(f"Здравствуй {
          name}! Здесь есть несколько режимов: Игра, Просмотр очков, Выход.\n")
    print("Вводите соответственно: 1, 2, 3. Окей?\n")

    # Сам процесс игры
    running = True  # Показатель того, нужно ли продолжать игру
    while running:  # Если игру нужно показывать:

        # Ввод режима
        mode = input("Введите номер режима: ")

        # Если режим "игра"
        if mode == "1":

            print("\nЧудесный выбор!")
            print(
                "\nВведите 2 положительных числа, между которыми будет загадано число:\n")

            # Счетчик для просмотра введенных чисел (а может и не чисел)
            counter_error = 0

            while counter_error != 1:  # Пока где-то ошибка

                global first_number
                global second_number

                # Просим ввести два числа, между которыми будет загадано число
                first_number = input("Введите первое число:")
                second_number = input("Введите второе число:")

                # Если введены только числа
                if first_number.isdigit() == True and second_number.isdigit() == True:

                    # Если первое число меньше второго (благоприятный исход)
                    if first_number < second_number:

                        counter_error = 1  # Прекращаем цикл спрашивания

                        # Все два числа преобразуем в числовой тип для дальнейшей работы с ними
                        first_number = int(first_number)
                        second_number = int(second_number)

                    # Если первое число больше второго (неблагоприятный исход)
                    else:

                        # Выводим пользователю напоминалку
                        print("\nПервое число должно быть меньше второго!\n")

                # Если введены НЕ только числа
                else:

                    # Выводим пользователю напоминалку
                    print("Нужно писать числа!")

            print()
            print("Числа приняты! Выбираю одно из их диапозона!")

            hidden_number = random.randint(
                first_number, second_number)  # Выьор числа

            time.sleep(2)  # Типо пару секунд компьютер думал

            # Компьютер оповещает пользователя о начале игры
            print("\nЯ выбрал число! Можете отгадывать!\n")

            counter_step = 0  # Счетчик шагов (для рекордов)

            while True:  # Будем продолжать до определенного момента

                # Я кое-что знаю об исключениях, так что я решил их использовать
                try:
                    # Запрашиваем ответ пользователя
                    answer = int(input())

                    # Каждый ход я буду добавлять по 1 в step_counter
                    # Чтобы в конце просмотреть кол-во ходов и занести надо ли в рекорды

                    # Если число больше загаданного
                    if answer > hidden_number:

                        print("\nЗагаданное мною число меньше твоего!\n")
                        counter_step += 1

                    # Если число меньше загаданного
                    elif answer < hidden_number:

                        print("\nЗагаданное мною число больше твоего!\n")
                        counter_step += 1

                    # Если пользователь попал в яблочко (отгадал число)
                    elif answer == hidden_number:

                        print("\nВоу, ты угадал мое число!\n")
                        time.sleep(1)
                        print(f"При этом ты потралил {
                              counter_step} попыток!\n")
                        time.sleep(1)

                        # Если кол-во ходов минимальное среди других
                        if mn > counter_step:

                            print(
                                "Кстати, это твой новый рекор по быстроте отгадывания!!!\n")
                            mn = counter_step  # Число ходов изменяется

                            # В эти данные запишем ту игру, когда пользователь поставил рекорд
                            first_number_fast = first_number
                            second_number_fast = second_number

                        # Если кол-во ходов максимальное среди других
                        if mx < counter_step:

                            print("Кстати, это твое самое долгое отгадывание!!!\n")
                            mx = counter_step  # Чсло ходов для этого рекорда меняется

                            # В эти данные запишем ту игру, когда пользователь поставил рекорд
                            first_number_long = first_number
                            second_number_long = second_number

                        counter_step = 0  # Обнуляем кол-во ходов для подсчета в будующих играх
                        counter_play += 1  # А к кол-ву игр добавляем одну

                        break  # Прекращаем цикл игры (не установки режима)

                # В случае, если игрок введет не число
                except:

                    print("\nНужно вводить числа, а не буквы!\n")

        # Если режим "Просмотр очков"
        elif mode == "2":

            # Оформление статистики
            print("\nВот твоя статистика:")
            decoration()
            print(f"Игр сыграно: {counter_play}")  # Показ кол-ва игр (всего)
            decoration()

            # Если еще ни одна игра не сыграна
            if counter_play == 0:
                print(f"Статистику получить невозможно, так как {
                      name} не сыграл ни одной игры!")
                print()
            # Если хоть одна игра, но сыграна
            else:
                print(f"Самое быстрое отгадывание (произошло в игре между числами {
                      first_number_fast} и {second_number_fast}): {mn} попыток!")
                decoration()
                print(f"Самое долгое отгадывание (произошло в игре между числами {
                      first_number_long} и {second_number_long}): {mx} попыток!")
                print()

        # Если режим "Выход"
        elif mode == "3":

            # Переспрашиваем пользователя (вдруг случайно?)
            print(f"\n{name} точно хочет выйти? Напишите только Да или Нет!")

            exit = input()  # Выходит пользователь или нет - здесь будет введен ответ!

            # Если пользователь все же хочет выйти
            if exit.lower() == 'да':

                print(f"\nНу ладно, всего доброго, {name}!\n")

                running = False  # Игра прекращается

            # Если все же ошибся
            elif exit.lower() == "нет":

                print("\nНу вот и хорошо ;)\n")

            # Если введет что-то кроме "Да" или "Нет"
            else:

                print("\nЯ не понял, что это? Возращаю на выбор режима!\n")


playing()  # Вызываем всю эту игру
```
