```python
# 1.1     ПЕРЕДЕЛАТЬ    другие списки     ПОДСЧЁТ ОТДЕЛЬНЫХ ДОМОВ ГОРОДА   ))
#  ОШИБКА  ОШИБКА ОШИБКА  когда здание внутри здания
'''  # ввод   nm = input().split(' ')
n,m = int(nm[0]), int(nm[1])
town = [*range(n)] 
for i in range(n):
    town[i] = input() '''

# готовые данные для примера
town = ['############',
        '#........###',
        '#.#####.##.#',
        '#.##.##.####',
        '#.#####.##.#',
        '#........###',
        '############']

n, m = len(town), len(town[0])


town_map = town[:]              # скопируеи данные в отдельную карту
for i in range(n):
    for j in range(m):          # и 'оцифруем' (вернее 'обуквим') её 'попиксельно'
        if town_map[i][j] == '#':  
            town_map[i] = town_map[i][:j] + str(chr(65+j)) + town_map[i][j+1:]
        if town_map[i][j] == '.':
            town_map[i] = town_map[i][:j] + str(' ') + town_map[i][j+1:]
    town_map[i] = town_map[i].split()
    print(town_map[i])
print()


# 'сканируем'
hause_cnt = 0               # количество домов
for i in range(n):
    if town_map[i-1] and i:    # предыдущий ряд есть    и это не первый ряд
        if town_map[i]:      # и этот ряд есть
            flag_last = ''
            for j in town_map[i]:    # перебор этих отрезков
                flag = False
                for k in town_map[i-1]:  # перебор верхних отрезков
                    if set(j).intersection(set(k)):
                        if flag_last == k:  # если пересеч с уже пересеч то лбъеденить отрезки 
                            ## print(town_map[i][town_map[i].index(j) - 1])
                            temp = town_map[i][town_map[i].index(j) - 1] + j
                            town_map[i][town_map[i].index(j) - 1] = ''
                            town_map[i][town_map[i].index(j)] = temp
                            ## print(town_map[i])
                        if flag:
                            # если несколько комнат сверху отнимаем (пбъединяем)
                            hause_cnt -= 1
                        flag = True
                        flag_last = k
                        ## print(set(j).intersection(set(k)))
                        ## print(flag_last)

                if flag:
                    pass
                else:
                    hause_cnt += 1   # если верхнего нет  засчитываем пока как дом

    else:               # предыдущего ряда нет    или он пустой
        hause_cnt += len(town_map[i])  # все отрезки засчитываем пока как дома

print()
print(hause_cnt)
```
