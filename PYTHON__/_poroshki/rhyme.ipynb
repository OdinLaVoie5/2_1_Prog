```python
import re

# -----------------------------
# ФОНЕТИЧЕСКАЯ НОРМАЛИЗАЦИЯ
# -----------------------------
VOWELS = "аеёиоуыэюя"
CONSONANTS = "бвгджзйклмнпрстфхцчшщ"

#  дб ^    ГРУППЫ РОДСТВЕННЫХ ЗВУКОВ   черновик  V
PHON_GROUPS = [    
    ("а", "о"),    
    ("ё", "е", "ю", "я"),         
    ("б", "п"),                 
    ("в", "ф"),                 
    ("г", "к", "х"),            
    ("д", "т"),                 
    ("ж", "ш"),                 
    ("л", "н", "м"),             
    ("й",),                     
    ("ч", "щ"),                 
    ("з", "с", "ц",),           
]

# делаем словарь: буква -> канонический символ группы
_CANON = {}
for idx, grp in enumerate(PHON_GROUPS):
    rep = grp[0]  # представитель группы   ?? 
    for ch in grp:
        _CANON[ch] = rep


def phon_norm_text(s: str) -> str:
    out = []
    for ch in s.lower():
        if ch in _CANON:
            out.append(_CANON[ch])
        elif ch in VOWELS or ch in CONSONANTS:
            out.append(ch)
        # прочее выбрасываем
    return "".join(out)

# -----------------------------
# ХВОСТ ДЛЯ РИФМЫ (твои правила)
# -----------------------------


def _last_vowel_idx(word: str) -> int | None:
    for i in range(len(word) - 1, -1, -1):
        if word[i] in VOWELS:
            return i
    return None


def _capture_from_last_vowel(word: str) -> str:
    """От последней гласной, но захватывая прилегающую слева согласную, если есть."""
    if not word:
        return ""
    i = _last_vowel_idx(word)
    if i is None:
        return word  # экзотика: слово без гласных
    start = i - 1 if i > 0 and word[i - 1] not in VOWELS else i
    return word[start:]


def get_rhyme_tail(line: str) -> str:
    """
    Берём последнее слово; если оно из 1 буквы:
      - если согласная: хвост от последней гласной ПРЕДпоследнего + эта согласная
      - если гласная: последняя согласная буква ПРЕДпоследнего (если есть),
                      иначе отступаем назад ещё на 1 символ, + эта гласная
    Иначе (обычное слово): от последней гласной (с захватом пред. согласной) до конца.
    """
    words = re.findall(r"[а-яё]+", line.lower())
    if not words:
        return ""
    last = words[-1]
    if len(last) > 1:
        return _capture_from_last_vowel(last)

    # --- однобуквенное слово ---
    if len(words) >= 2:
        prev = words[-2]

        if last in VOWELS:
            # ищем последнюю звучащую согласную в конце предыдущего слова
            i = len(prev) - 1
            while i >= 0 and (prev[i] in VOWELS or prev[i] in ("ь", "ъ")):
                i -= 1
            if i >= 0:
                return prev[i:] + last  # напр. "ночь а" → "чьа"
            else:
                return prev + last      # fallback: если ничего не нашли
        else:
            # "баран к" → "ранк"
            base = _capture_from_last_vowel(prev)
            return base + last
    return last

# -----------------------------
# ЛЕВЕНШТЕЙН
# -----------------------------


def levenshtein(a: str, b: str) -> int:
    if a == b:
        return 0
    if len(a) == 0:
        return len(b)
    if len(b) == 0:
        return len(a)
    dp = list(range(len(b) + 1))
    for i, ca in enumerate(a, start=1):
        prev, dp[0] = dp[0], i
        for j, cb in enumerate(b, start=1):
            cur = dp[j]
            cost = 0 if ca == cb else 1
            dp[j] = min(
                dp[j] + 1,       # вставка
                dp[j - 1] + 1,   # удаление
                prev + cost      # замена
            )
            prev = cur
    return dp[-1]

# -----------------------------
# ОЦЕНКА РИФМЫ И АССОНАНСА
# -----------------------------


def rhyme_soft(line2: str, line4: str,
               hard_min_suffix: int = 2,
               max_lev_tail: int = 1,
               assonance_rel_threshold: float = 0.35):
    """
    Возвращает (ok: bool, method: str, info: dict).
    1) Рифма по хвостам (фонетнорм) — точное равенство / общий суффикс >= hard_min_suffix / Левенштейн <= max_lev_tail
    2) Если не ок — ассонанс: берём |line4| последних символов из line2 (по буквам),
       считаем Левенштейна по фонетически нормализованным срезам и нормируем на длину.
    """
    t2 = phon_norm_text(get_rhyme_tail(line2))
    t4 = phon_norm_text(get_rhyme_tail(line4))

    # 1) мягкая рифма (фонетические хвосты)
    if t2 and t4:
        if t2 == t4:
            return True, "rhyme_eq", {"t2": t2, "t4": t4}
        if len(t2) >= hard_min_suffix and len(t4) >= hard_min_suffix:
            if t2[-hard_min_suffix:] == t4[-hard_min_suffix:]:
                return True, f"rhyme_suffix{hard_min_suffix}", {"t2": t2, "t4": t4}
        if levenshtein(t2, t4) <= max_lev_tail:
            return True, "rhyme_lev", {"t2": t2, "t4": t4, "lev": levenshtein(t2, t4)}

    # 2) ассонанс: выравниваем длину по длине строки 4
    # считаем по буквам (только а-яё), нормализуем фонетически
    w2 = re.findall(r"[а-яё]", line2.lower())
    w4 = re.findall(r"[а-яё]", line4.lower())
    if not w2 or not w4:
        return False, "no_letters", {}

    n = len(w4)
    tail2 = "".join(w2[-n:]) if len(w2) >= n else "".join(w2)
    tail4 = "".join(w4[-n:])

    p2 = phon_norm_text(tail2)
    p4 = phon_norm_text(tail4)
    if not p2 or not p4:
        return False, "no_phon", {"tail2": tail2, "tail4": tail4}

    d = levenshtein(p2, p4)
    rel = d / max(len(p2), len(p4))
    ok = rel <= assonance_rel_threshold
    return ok, "assonance", {"p2": p2, "p4": p4, "lev": d, "rel": rel}

# -----------------------------
# ТЕСТЫ И ПРИМЕРЫ
# ----------------------------- 
examples = [
    ("терпи", "оркестр"),   # кестр
    ("дрова н", "баран к"),   # ранк
]

for line2, line4 in examples:
    ok, method, info = rhyme_soft(line2, line4)
    status = "OK" if ok else "NO"
    print(f"{status:3} {method:20} {line2!r:15} ~ {line4!r:15}  {info}")
```

```python
def extract_rhyme_tail(line):
    vowels = "аеёиоуыэюя"
    silent = "ьъ"
    words = line.strip().split()
    if not words:
        return ""

    last = words[-1]
    penultimate = words[-2] if len(words) > 1 else ""

    def last_vowel_index(word):
        for i in range(len(word)-1, -1, -1):
            if word[i] in vowels:
                return i
        return -1 

    def adjust_tail(word, idx):
        # если последняя буква — гласная или ъ/ь, захватываем ещё одну слева
        tail = word[idx:]
        if word[-1] in vowels + silent and idx > 0:
            tail = word[idx-1:]
        return tail

    # Case 1: обычное слово в конце
    if len(last) > 1:
        idx = last_vowel_index(last)
        if idx == -1:
            return last
        return adjust_tail(last, idx)

    # Case 2: однобуквенное слово — согласная
    if last not in vowels:
        idx = last_vowel_index(penultimate)
        if idx == -1:
            return last
        tail = adjust_tail(penultimate, idx)
        return tail + last

    # Case 3: однобуквенное слово — гласная
    if penultimate:
        tail = penultimate[-1]
        if tail in vowels + silent and len(penultimate) > 1:
            tail = penultimate[-2] + tail
        return tail + last

    return last

# 🔍 Примеры
examples = [
    "оркестр",
    "баран к",
    "весна в",
    "дочек а",
    "пыль",
    "мрак и",
    "тень у",
    "сна а",
    "ин ян"

]

for line in examples:
    print(f"{line!r} → {extract_rhyme_tail(line)!r}")
```

```python
import re

VOWELS = "аеёиоуыэюя"


def get_rhyme_tail(line: str) -> str:
    words = re.findall(r"[а-яё]+", line.lower())
    if not words:
        return ""

    last = words[-1]

    # 1) обычное слово (больше 1 буквы)
    if len(last) > 1:
        for i in range(len(last)-1, -1, -1):
            if last[i] in VOWELS:
                start = i-1 if i > 0 and last[i-1] not in VOWELS else i
                return last[start:]
        return last

    # 2) однобуквенное слово
    if len(words) >= 2:
        prev = words[-2]
        if last in VOWELS:
            return prev[-1:] + last
        else:
            for i in range(len(prev)-1, -1, -1):
                if prev[i] in VOWELS:
                    start = i-1 if i > 0 and prev[i-1] not in VOWELS else i
                    return prev[start:] + last
            return prev + last
    return last

# см выше пакет levenshtein
def levenshtein(a: str, b: str) -> int:
    """Минимальное количество операций вставки/удаления/замены"""
    dp = [
        [i+j if i*j == 0 else 0 for j in range(len(b)+1)] for i in range(len(a)+1)]
    for i in range(1, len(a)+1):
        for j in range(1, len(b)+1):
            cost = 0 if a[i-1] == b[j-1] else 1
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
    return dp[-1][-1]


def rhymes(line1: str, line2: str, soft: bool = True) -> bool:
    t1, t2 = get_rhyme_tail(line1), get_rhyme_tail(line2)
    if not soft:
        return t1 == t2

    # мягкая проверка
    if t1 == t2:
        return True
    if t1[-2:] == t2[-2:]:   # совпадение 2 букв
        return True
    if t1[-3:] == t2[-3:]:   # совпадение 3 букв
        return True
    if levenshtein(t1, t2) <= 1:  # почти одинаковые
        return True
    return False


# 🔹 тесты
print(get_rhyme_tail("оркестр"))   # кестр
print(get_rhyme_tail("баран к"))   # ранк
print(get_rhyme_tail("весна в"))   # снав
print(get_rhyme_tail("дочек а"))   # чека
print(get_rhyme_tail("ночь я"))   # чека

# True (кестр / ист → нет, но 2 буквы совпадут)
print(rhymes("оркестр", "регист"))
print(rhymes("баран к", "караул в"))   
```
