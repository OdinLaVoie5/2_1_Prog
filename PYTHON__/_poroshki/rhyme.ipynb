```python
import re

# -----------------------------
# –§–û–ù–ï–¢–ò–ß–ï–°–ö–ê–Ø –ù–û–†–ú–ê–õ–ò–ó–ê–¶–ò–Ø
# -----------------------------
VOWELS = "–∞–µ—ë–∏–æ—É—ã—ç—é—è"
CONSONANTS = "–±–≤–≥–¥–∂–∑–π–∫–ª–º–Ω–ø—Ä—Å—Ç—Ñ—Ö—Ü—á—à—â"

#  –¥–± ^    –ì–†–£–ü–ü–´ –†–û–î–°–¢–í–ï–ù–ù–´–• –ó–í–£–ö–û–í   —á–µ—Ä–Ω–æ–≤–∏–∫  V
PHON_GROUPS = [    
    ("–∞", "–æ"),    
    ("—ë", "–µ", "—é", "—è"),         
    ("–±", "–ø"),                 
    ("–≤", "—Ñ"),                 
    ("–≥", "–∫", "—Ö"),            
    ("–¥", "—Ç"),                 
    ("–∂", "—à"),                 
    ("–ª", "–Ω", "–º"),             
    ("–π",),                     
    ("—á", "—â"),                 
    ("–∑", "—Å", "—Ü",),           
]

# –¥–µ–ª–∞–µ–º —Å–ª–æ–≤–∞—Ä—å: –±—É–∫–≤–∞ -> –∫–∞–Ω–æ–Ω–∏—á–µ—Å–∫–∏–π —Å–∏–º–≤–æ–ª –≥—Ä—É–ø–ø—ã
_CANON = {}
for idx, grp in enumerate(PHON_GROUPS):
    rep = grp[0]  # –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª—å –≥—Ä—É–ø–ø—ã   ?? 
    for ch in grp:
        _CANON[ch] = rep


def phon_norm_text(s: str) -> str:
    out = []
    for ch in s.lower():
        if ch in _CANON:
            out.append(_CANON[ch])
        elif ch in VOWELS or ch in CONSONANTS:
            out.append(ch)
        # –ø—Ä–æ—á–µ–µ –≤—ã–±—Ä–∞—Å—ã–≤–∞–µ–º
    return "".join(out)

# -----------------------------
# –•–í–û–°–¢ –î–õ–Ø –†–ò–§–ú–´ (—Ç–≤–æ–∏ –ø—Ä–∞–≤–∏–ª–∞)
# -----------------------------


def _last_vowel_idx(word: str) -> int | None:
    for i in range(len(word) - 1, -1, -1):
        if word[i] in VOWELS:
            return i
    return None


def _capture_from_last_vowel(word: str) -> str:
    """–û—Ç –ø–æ—Å–ª–µ–¥–Ω–µ–π –≥–ª–∞—Å–Ω–æ–π, –Ω–æ –∑–∞—Ö–≤–∞—Ç—ã–≤–∞—è –ø—Ä–∏–ª–µ–≥–∞—é—â—É—é —Å–ª–µ–≤–∞ —Å–æ–≥–ª–∞—Å–Ω—É—é, –µ—Å–ª–∏ –µ—Å—Ç—å."""
    if not word:
        return ""
    i = _last_vowel_idx(word)
    if i is None:
        return word  # —ç–∫–∑–æ—Ç–∏–∫–∞: —Å–ª–æ–≤–æ –±–µ–∑ –≥–ª–∞—Å–Ω—ã—Ö
    start = i - 1 if i > 0 and word[i - 1] not in VOWELS else i
    return word[start:]


def get_rhyme_tail(line: str) -> str:
    """
    –ë–µ—Ä—ë–º –ø–æ—Å–ª–µ–¥–Ω–µ–µ —Å–ª–æ–≤–æ; –µ—Å–ª–∏ –æ–Ω–æ –∏–∑ 1 –±—É–∫–≤—ã:
      - –µ—Å–ª–∏ —Å–æ–≥–ª–∞—Å–Ω–∞—è: —Ö–≤–æ—Å—Ç –æ—Ç –ø–æ—Å–ª–µ–¥–Ω–µ–π –≥–ª–∞—Å–Ω–æ–π –ü–†–ï–î–ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ + —ç—Ç–∞ —Å–æ–≥–ª–∞—Å–Ω–∞—è
      - –µ—Å–ª–∏ –≥–ª–∞—Å–Ω–∞—è: –ø–æ—Å–ª–µ–¥–Ω—è—è —Å–æ–≥–ª–∞—Å–Ω–∞—è –±—É–∫–≤–∞ –ü–†–ï–î–ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ (–µ—Å–ª–∏ –µ—Å—Ç—å),
                      –∏–Ω–∞—á–µ –æ—Ç—Å—Ç—É–ø–∞–µ–º –Ω–∞–∑–∞–¥ –µ—â—ë –Ω–∞ 1 —Å–∏–º–≤–æ–ª, + —ç—Ç–∞ –≥–ª–∞—Å–Ω–∞—è
    –ò–Ω–∞—á–µ (–æ–±—ã—á–Ω–æ–µ —Å–ª–æ–≤–æ): –æ—Ç –ø–æ—Å–ª–µ–¥–Ω–µ–π –≥–ª–∞—Å–Ω–æ–π (—Å –∑–∞—Ö–≤–∞—Ç–æ–º –ø—Ä–µ–¥. —Å–æ–≥–ª–∞—Å–Ω–æ–π) –¥–æ –∫–æ–Ω—Ü–∞.
    """
    words = re.findall(r"[–∞-—è—ë]+", line.lower())
    if not words:
        return ""
    last = words[-1]
    if len(last) > 1:
        return _capture_from_last_vowel(last)

    # --- –æ–¥–Ω–æ–±—É–∫–≤–µ–Ω–Ω–æ–µ —Å–ª–æ–≤–æ ---
    if len(words) >= 2:
        prev = words[-2]

        if last in VOWELS:
            # –∏—â–µ–º –ø–æ—Å–ª–µ–¥–Ω—é—é –∑–≤—É—á–∞—â—É—é —Å–æ–≥–ª–∞—Å–Ω—É—é –≤ –∫–æ–Ω—Ü–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Å–ª–æ–≤–∞
            i = len(prev) - 1
            while i >= 0 and (prev[i] in VOWELS or prev[i] in ("—å", "—ä")):
                i -= 1
            if i >= 0:
                return prev[i:] + last  # –Ω–∞–ø—Ä. "–Ω–æ—á—å –∞" ‚Üí "—á—å–∞"
            else:
                return prev + last      # fallback: –µ—Å–ª–∏ –Ω–∏—á–µ–≥–æ –Ω–µ –Ω–∞—à–ª–∏
        else:
            # "–±–∞—Ä–∞–Ω –∫" ‚Üí "—Ä–∞–Ω–∫"
            base = _capture_from_last_vowel(prev)
            return base + last
    return last

# -----------------------------
# –õ–ï–í–ï–ù–®–¢–ï–ô–ù
# -----------------------------


def levenshtein(a: str, b: str) -> int:
    if a == b:
        return 0
    if len(a) == 0:
        return len(b)
    if len(b) == 0:
        return len(a)
    dp = list(range(len(b) + 1))
    for i, ca in enumerate(a, start=1):
        prev, dp[0] = dp[0], i
        for j, cb in enumerate(b, start=1):
            cur = dp[j]
            cost = 0 if ca == cb else 1
            dp[j] = min(
                dp[j] + 1,       # –≤—Å—Ç–∞–≤–∫–∞
                dp[j - 1] + 1,   # —É–¥–∞–ª–µ–Ω–∏–µ
                prev + cost      # –∑–∞–º–µ–Ω–∞
            )
            prev = cur
    return dp[-1]

# -----------------------------
# –û–¶–ï–ù–ö–ê –†–ò–§–ú–´ –ò –ê–°–°–û–ù–ê–ù–°–ê
# -----------------------------


def rhyme_soft(line2: str, line4: str,
               hard_min_suffix: int = 2,
               max_lev_tail: int = 1,
               assonance_rel_threshold: float = 0.35):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç (ok: bool, method: str, info: dict).
    1) –†–∏—Ñ–º–∞ –ø–æ —Ö–≤–æ—Å—Ç–∞–º (—Ñ–æ–Ω–µ—Ç–Ω–æ—Ä–º) ‚Äî —Ç–æ—á–Ω–æ–µ —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ / –æ–±—â–∏–π —Å—É—Ñ—Ñ–∏–∫—Å >= hard_min_suffix / –õ–µ–≤–µ–Ω—à—Ç–µ–π–Ω <= max_lev_tail
    2) –ï—Å–ª–∏ –Ω–µ –æ–∫ ‚Äî –∞—Å—Å–æ–Ω–∞–Ω—Å: –±–µ—Ä—ë–º |line4| –ø–æ—Å–ª–µ–¥–Ω–∏—Ö —Å–∏–º–≤–æ–ª–æ–≤ –∏–∑ line2 (–ø–æ –±—É–∫–≤–∞–º),
       —Å—á–∏—Ç–∞–µ–º –õ–µ–≤–µ–Ω—à—Ç–µ–π–Ω–∞ –ø–æ —Ñ–æ–Ω–µ—Ç–∏—á–µ—Å–∫–∏ –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–º —Å—Ä–µ–∑–∞–º –∏ –Ω–æ—Ä–º–∏—Ä—É–µ–º –Ω–∞ –¥–ª–∏–Ω—É.
    """
    t2 = phon_norm_text(get_rhyme_tail(line2))
    t4 = phon_norm_text(get_rhyme_tail(line4))

    # 1) –º—è–≥–∫–∞—è —Ä–∏—Ñ–º–∞ (—Ñ–æ–Ω–µ—Ç–∏—á–µ—Å–∫–∏–µ —Ö–≤–æ—Å—Ç—ã)
    if t2 and t4:
        if t2 == t4:
            return True, "rhyme_eq", {"t2": t2, "t4": t4}
        if len(t2) >= hard_min_suffix and len(t4) >= hard_min_suffix:
            if t2[-hard_min_suffix:] == t4[-hard_min_suffix:]:
                return True, f"rhyme_suffix{hard_min_suffix}", {"t2": t2, "t4": t4}
        if levenshtein(t2, t4) <= max_lev_tail:
            return True, "rhyme_lev", {"t2": t2, "t4": t4, "lev": levenshtein(t2, t4)}

    # 2) –∞—Å—Å–æ–Ω–∞–Ω—Å: –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–µ–º –¥–ª–∏–Ω—É –ø–æ –¥–ª–∏–Ω–µ —Å—Ç—Ä–æ–∫–∏ 4
    # —Å—á–∏—Ç–∞–µ–º –ø–æ –±—É–∫–≤–∞–º (—Ç–æ–ª—å–∫–æ –∞-—è—ë), –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º —Ñ–æ–Ω–µ—Ç–∏—á–µ—Å–∫–∏
    w2 = re.findall(r"[–∞-—è—ë]", line2.lower())
    w4 = re.findall(r"[–∞-—è—ë]", line4.lower())
    if not w2 or not w4:
        return False, "no_letters", {}

    n = len(w4)
    tail2 = "".join(w2[-n:]) if len(w2) >= n else "".join(w2)
    tail4 = "".join(w4[-n:])

    p2 = phon_norm_text(tail2)
    p4 = phon_norm_text(tail4)
    if not p2 or not p4:
        return False, "no_phon", {"tail2": tail2, "tail4": tail4}

    d = levenshtein(p2, p4)
    rel = d / max(len(p2), len(p4))
    ok = rel <= assonance_rel_threshold
    return ok, "assonance", {"p2": p2, "p4": p4, "lev": d, "rel": rel}

# -----------------------------
# –¢–ï–°–¢–´ –ò –ü–†–ò–ú–ï–†–´
# ----------------------------- 
examples = [
    ("—Ç–µ—Ä–ø–∏", "–æ—Ä–∫–µ—Å—Ç—Ä"),   # –∫–µ—Å—Ç—Ä
    ("–¥—Ä–æ–≤–∞ –Ω", "–±–∞—Ä–∞–Ω –∫"),   # —Ä–∞–Ω–∫
]

for line2, line4 in examples:
    ok, method, info = rhyme_soft(line2, line4)
    status = "OK" if ok else "NO"
    print(f"{status:3} {method:20} {line2!r:15} ~ {line4!r:15}  {info}")
```

```python
def extract_rhyme_tail(line):
    vowels = "–∞–µ—ë–∏–æ—É—ã—ç—é—è"
    silent = "—å—ä"
    words = line.strip().split()
    if not words:
        return ""

    last = words[-1]
    penultimate = words[-2] if len(words) > 1 else ""

    def last_vowel_index(word):
        for i in range(len(word)-1, -1, -1):
            if word[i] in vowels:
                return i
        return -1 

    def adjust_tail(word, idx):
        # –µ—Å–ª–∏ –ø–æ—Å–ª–µ–¥–Ω—è—è –±—É–∫–≤–∞ ‚Äî –≥–ª–∞—Å–Ω–∞—è –∏–ª–∏ —ä/—å, –∑–∞—Ö–≤–∞—Ç—ã–≤–∞–µ–º –µ—â—ë –æ–¥–Ω—É —Å–ª–µ–≤–∞
        tail = word[idx:]
        if word[-1] in vowels + silent and idx > 0:
            tail = word[idx-1:]
        return tail

    # Case 1: –æ–±—ã—á–Ω–æ–µ —Å–ª–æ–≤–æ –≤ –∫–æ–Ω—Ü–µ
    if len(last) > 1:
        idx = last_vowel_index(last)
        if idx == -1:
            return last
        return adjust_tail(last, idx)

    # Case 2: –æ–¥–Ω–æ–±—É–∫–≤–µ–Ω–Ω–æ–µ —Å–ª–æ–≤–æ ‚Äî —Å–æ–≥–ª–∞—Å–Ω–∞—è
    if last not in vowels:
        idx = last_vowel_index(penultimate)
        if idx == -1:
            return last
        tail = adjust_tail(penultimate, idx)
        return tail + last

    # Case 3: –æ–¥–Ω–æ–±—É–∫–≤–µ–Ω–Ω–æ–µ —Å–ª–æ–≤–æ ‚Äî –≥–ª–∞—Å–Ω–∞—è
    if penultimate:
        tail = penultimate[-1]
        if tail in vowels + silent and len(penultimate) > 1:
            tail = penultimate[-2] + tail
        return tail + last

    return last

# üîç –ü—Ä–∏–º–µ—Ä—ã
examples = [
    "–æ—Ä–∫–µ—Å—Ç—Ä",
    "–±–∞—Ä–∞–Ω –∫",
    "–≤–µ—Å–Ω–∞ –≤",
    "–¥–æ—á–µ–∫ –∞",
    "–ø—ã–ª—å",
    "–º—Ä–∞–∫ –∏",
    "—Ç–µ–Ω—å —É",
    "—Å–Ω–∞ –∞",
    "–∏–Ω —è–Ω"

]

for line in examples:
    print(f"{line!r} ‚Üí {extract_rhyme_tail(line)!r}")
```

```python
import re

VOWELS = "–∞–µ—ë–∏–æ—É—ã—ç—é—è"


def get_rhyme_tail(line: str) -> str:
    words = re.findall(r"[–∞-—è—ë]+", line.lower())
    if not words:
        return ""

    last = words[-1]

    # 1) –æ–±—ã—á–Ω–æ–µ —Å–ª–æ–≤–æ (–±–æ–ª—å—à–µ 1 –±—É–∫–≤—ã)
    if len(last) > 1:
        for i in range(len(last)-1, -1, -1):
            if last[i] in VOWELS:
                start = i-1 if i > 0 and last[i-1] not in VOWELS else i
                return last[start:]
        return last

    # 2) –æ–¥–Ω–æ–±—É–∫–≤–µ–Ω–Ω–æ–µ —Å–ª–æ–≤–æ
    if len(words) >= 2:
        prev = words[-2]
        if last in VOWELS:
            return prev[-1:] + last
        else:
            for i in range(len(prev)-1, -1, -1):
                if prev[i] in VOWELS:
                    start = i-1 if i > 0 and prev[i-1] not in VOWELS else i
                    return prev[start:] + last
            return prev + last
    return last

# —Å–º –≤—ã—à–µ –ø–∞–∫–µ—Ç levenshtein
def levenshtein(a: str, b: str) -> int:
    """–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –≤—Å—Ç–∞–≤–∫–∏/—É–¥–∞–ª–µ–Ω–∏—è/–∑–∞–º–µ–Ω—ã"""
    dp = [
        [i+j if i*j == 0 else 0 for j in range(len(b)+1)] for i in range(len(a)+1)]
    for i in range(1, len(a)+1):
        for j in range(1, len(b)+1):
            cost = 0 if a[i-1] == b[j-1] else 1
            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )
    return dp[-1][-1]


def rhymes(line1: str, line2: str, soft: bool = True) -> bool:
    t1, t2 = get_rhyme_tail(line1), get_rhyme_tail(line2)
    if not soft:
        return t1 == t2

    # –º—è–≥–∫–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞
    if t1 == t2:
        return True
    if t1[-2:] == t2[-2:]:   # —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ 2 –±—É–∫–≤
        return True
    if t1[-3:] == t2[-3:]:   # —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ 3 –±—É–∫–≤
        return True
    if levenshtein(t1, t2) <= 1:  # –ø–æ—á—Ç–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ
        return True
    return False


# üîπ —Ç–µ—Å—Ç—ã
print(get_rhyme_tail("–æ—Ä–∫–µ—Å—Ç—Ä"))   # –∫–µ—Å—Ç—Ä
print(get_rhyme_tail("–±–∞—Ä–∞–Ω –∫"))   # —Ä–∞–Ω–∫
print(get_rhyme_tail("–≤–µ—Å–Ω–∞ –≤"))   # —Å–Ω–∞–≤
print(get_rhyme_tail("–¥–æ—á–µ–∫ –∞"))   # —á–µ–∫–∞
print(get_rhyme_tail("–Ω–æ—á—å —è"))   # —á–µ–∫–∞

# True (–∫–µ—Å—Ç—Ä / –∏—Å—Ç ‚Üí –Ω–µ—Ç, –Ω–æ 2 –±—É–∫–≤—ã —Å–æ–≤–ø–∞–¥—É—Ç)
print(rhymes("–æ—Ä–∫–µ—Å—Ç—Ä", "—Ä–µ–≥–∏—Å—Ç"))
print(rhymes("–±–∞—Ä–∞–Ω –∫", "–∫–∞—Ä–∞—É–ª –≤"))   
```
