###  plot

```python
import matplotlib.pyplot as plt    # основная ф-ция  pyplot
import numpy as np                 # часто вместе с numpy
xpoints = np.array([0, 6])   # массивы 
ypoints = np.array([0, 250])
plt.plot(xpoints, ypoints)  # можно просто списки plt.plot([2,8], [3,10])

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}
# заголовок шрифт (цвет, размер)  выравнивать (умолч  center ) 
plt.title("Sports Watch Data", fontdict=font1, loc='right')  
plt.xlabel("Average Pulse", fontdict=font2, loc='left')  # подписи к осям
plt.ylabel("Calorie Burnage", fontdict=font2, loc='bottom')   # top

# plt.grid(axis='x')     plt.grid() сетка     axis='y' (умолч 'both' оба)
plt.grid(c='g', ls=':', lw='0.3')     # оформление см ниже

plt.show()
```

```python
## можно один список    ((второй по умолч [0,1,2,3 ...]))
#  (ОБЩЕЕ) оформление (в конце)
#  marker='o' (стиль маркера)   ls=':' (стиль линии)  c='g'  (цвет)     lw=15   (ширина линии)
#  ms=12 (размер маркера)    mfc='b' (цвет маркера внут)    mec = 'r' (цвет маркера внеш)
plt.plot([2, 6, 7, 6, 7, 10], marker='o', c='k', mec='#dec185', mfc='c', ls=':', lw=3.5, ms=17)
plt.plot([0, 5], [1,3])       #  можно несколько plot 
plt.plot([1,3], [0, 5])
plt.show()
```

```python
##  можно несколько пар списков (одинаковой длины)  (+ одиночные)
#  (ОТДЕЛЬНОЕ) оформление (после массивов (пары или одного)) -
#   '*:b'     'стиль маркера   стиль линии   цвет'
#   '*b'  или  '*'      если опущен стиль линии - значит без линии
#   'b'   но если указан только цвет - то линия рисуется (сплошная)
#         по умолчанию - сплошная линия без маркеров
plt.plot([2, 3, 4, 5, 6], [2, 6, 4, 6, 2], [10, 3.5, 1], '-.g', [1, 3, 4, 8], [
         2, 7, 6, 10], '*--r',  [5, 8, 7.6, 9.7, 8.5, 10], 'D:b',  [1, 3.5, 10],    mec='k', mfc='c', lw=2.5, ms=7)
#  сочетание отдельного и общего оформления - в случае конфликта пишет типа ошибка
plt.show()
```

### scatter

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
plt.scatter(x, y, c='hotpink')            # оформление см выше 

x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91])
plt.scatter(x, y)

plt.show()
```

```python
#  раскрасим каждую точку   (массив цветов)
colors = np.array(["red", "green", "blue", "yellow", "pink", "black",
                  "orange", "purple", "beige", "brown", "gray", "cyan", "magenta"])
x = np.array([2, 2, 8, 1, 15, 8, 12, 9, 7, 3, 11, 4, 7])
y = np.array([100, 105, 84, 105, 90, 99, 90, 95, 94, 100, 79, 112, 91])
plt.scatter(x, y, c=colors)

plt.show()
```

```python
plt.scatter([5, 7, 18, 27, 2], [44, 85, 34, 78, 92], c=['red', 'green', 'blue', 'yellow', 'pink'])
plt.show()
```

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
mycolors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60,
                    70, 80, 90, 100])      # (массив цветов)
mysizes = np.array([20, 50, 100, 200, 500, 1000, 60, 90, 10,
                   300, 600, 800, 75])   # и  (массив размеров)

# размер и встроенная цветовая карта н/р viridis (много разных)  и прозрачность
plt.scatter(x, y, s=mysizes, c=mycolors, cmap='viridis', alpha=0.5)
plt.colorbar()                   # показать цветовую карту 
plt.show()
```

```python
x = np.random.randint(100, size=(100))  # ПРИМЕР С СЛУЧ МАССИВОМ
y = np.random.randint(100, size=(100))
mycolors = np.random.randint(100, size=(100))
mysizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=mycolors, s=mysizes, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

### 	bar (бар (стержни) ?),     hist (гистограмма),      pie (круговые диаграммы)

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 8, 3, 10])

plt.bar(x, y, color="#4CAF50", width=0.5)  # типо гистограмма
# plt.barh(x, y, height=0.1, color="red")  # то же горизонтально

plt.show()
```

```python
x = np.random.normal(170, 10, 250)
plt.subplot(1, 2, 1) 
plt.hist(x, color='red')     # , width=5

plt.subplot(1, 2, 2)
plt.hist([1, 2, 3, 1, 7, 4, 5, 2, 3, 1, 5, 7, 4, 8, 9, 6, 4, 3, 1, 2, 5, 6, 7, 8, 9, 10], color='pink')
# (( по горизонтали - значения    по вертикали - кол-во повторов ))

plt.show()
```

```python
y = np.array([7, 25, 25, 15, 35]) 
mycolors = ["black", "hotpink", "b", "#4CAF50", "red"]

plt.pie(y, labels=["Я", "Ты", "Он", "Она", "Вместе"], startangle=90, explode=[0.2, 0.1, 0.2, -0.1, 0.2], shadow=True, colors=mycolors)
plt.legend(title="Four Fruits:")

plt.show()
```
