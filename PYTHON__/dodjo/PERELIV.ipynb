### Исполнитель «Водолей», 
пожалуй, один из самых известных. Его работа тесно связана с 
алгоритмом Евклида. По соотношению Безу (свойство НОД №9) найдутся такие x и y, что
НОД(А, B) = xA + yB (x и y могут быть отрицательными). 
Если требуемый объем не кратен НОД(А, B) – решения нет. Если кратен, то решение есть и 
мы рано или поздно получим его (по условию это как раз гарантируется).
Возьмем для определенности x положительным, а y отрицательным. Тогда x будет 
соответствовать количеству «вливаний» в систему объемов А , а y – выливанию объемов 
13
B. (Алгоритм прост: наполняем А, выливаем его в В, как только B наполняется –
выливаем B). 
Конечно, необходимо учесть случай, когда отмеряемое количество больше объема 
сосуда, как во втором примере. Это можно проверить сразу и вывести Impossible

```python
# 1.4 !!!!!  ПЕРЕЛИВАНИЕ СОСУДОВ (водолей)        циклы  убрал
# (( каждый цикл while содержит 2 операции (A>B и (>A или >B)  ИЛИ  B>A и (>A или >B))
# число циклов максимум A+B        ИТОГ (A+B)*2   
# ИТОГ на практике  ((A+B-2)*2    /2  ?)          ))

from math import gcd

A, B, N = 17, 18, 7    # int(input()), int(input()), int(input())

if N == 0: print('')      # надо ли? 
elif any([A<0, B<0, N<0]):print('Impossible')   # надо ли?

elif A == N:        # поменять порядок? ↡↡
    print('>A')
elif B == N:
    print('>B')
elif (N > max(A, B)) or (N % gcd(A, B)): 
    print('Impossible')

else:
    tempA, tempB = A, B     # ДЛЯ ЗАМЕНЫ В КОНЦЕ если порядок  A>B
    if A > B: A, B = B, A

    cur_a1, cur_b1 = A, 0    # направление 1 (наливаем из A в B)
    resA = '>A'
    cur_a2, cur_b2 = 0, B  # направление 2  (черпаем из B в A)
    resB = '>B'

    # cnt = 0
    while True:
        
        # направление 1 (наливаем из A в B)
        if B - cur_b1 > A:
            cur_b1 += cur_a1
            resA += ' A>B'
            if cur_b1 == N:   # нашёл
                res = resA
                break
            cur_a1 = A
            resA += ' >A'
        else:
            cur_a1 = A - (B - cur_b1)
            resA += ' A>B'
            if cur_a1 == N:   # нашёл
                res = resA
                break
            cur_b1 = 0
            resA += ' B>'

        # направление 2  (черпаем из B в A)
        if cur_b2 > A - cur_a2:
            cur_b2 -= (A - cur_a2)
            resB += ' B>A'
            if cur_b2 == N:   # нашёл
                res = resB
                break
            cur_a2 = 0
            resB += ' A>'
        else:
            cur_a2 += cur_b2
            resB += ' B>A'
            if cur_a2 == N:   # нашёл
                res = resB
                break
            cur_b2 = B
            resB += ' >B'

    if tempA > tempB:      # (костыль) меняем A и B местами
        res = res.replace('A', 'C')
        res = res.replace('B', 'A')
        res = res.replace('C', 'B')

    print(res.replace(' ', '\n'))   # (оформление в столбик)
    # print(cnt)
    # print(res)                    # (оформление в линию)
```

```python
#--- чуж       рекурсия 
def water_jug_problem(A, B, N):
    path = []
    visited_states = set()

    def dfs(current_state):
        if current_state[0] == N or current_state[1] == N:
            return True

        visited_states.add(current_state)

        # Наполнить сосуд A
        if current_state[0] < A and (A, current_state[1]) not in visited_states:
            path.append(">A")
            if dfs((A, current_state[1])):
                return True
            path.pop()

        # Наполнить сосуд B
        if current_state[1] < B and (current_state[0], B) not in visited_states:
            path.append(">B")
            if dfs((current_state[0], B)):
                return True
            path.pop()

        # Вылить из сосуда A
        if current_state[0] > 0 and (0, current_state[1]) not in visited_states:
            path.append("A>")
            if dfs((0, current_state[1])):
                return True
            path.pop()

        # Вылить из сосуда B
        if current_state[1] > 0 and (current_state[0], 0) not in visited_states:
            path.append("B>")
            if dfs((current_state[0], 0)):
                return True
            path.pop()

        # Переливание из A в B
        if current_state[0] > 0 and current_state[1] < B:
            transfer = min(current_state[0], B - current_state[1])
            new_state = (current_state[0] - transfer,
                         current_state[1] + transfer)
            if new_state not in visited_states:
                path.append("A>B")
                if dfs(new_state):
                    return True
                path.pop()

        # Переливание из B в A
        if current_state[1] > 0 and current_state[0] < A:
            transfer = min(current_state[1], A - current_state[0])
            new_state = (current_state[0] + transfer,
                         current_state[1] - transfer)
            if new_state not in visited_states:
                path.append("B>A")
                if dfs(new_state):
                    return True
                path.pop()

        return False

    if dfs((0, 0)):
        return path
    else:
        return "Impossible"


# Пример использования
A, B, N = 3, 5, 1
result = water_jug_problem(A, B, N)
print(result)
```

```python
### 2.1  для любого k   (список) 

def joseph_countdown_list(n, k):
    live_list = list(range(1, n+1)) 
    dead_list = []
    cnt = i = 0

    while live_list:
        cnt += 1
        if cnt % k == 0:            # жребий выпал
            dead_list.append(live_list[i]) 
            live_list[i] = 0        # перекинуть выбывших и занулить их
            
        i += 1
        if i == len(live_list):      # цикл закончился - обновить
            while 0 in live_list:
                live_list.remove(0)  # очистить следы выбывших
            i = 0                    # обнулить индикатор

    return dead_list


n, k = 14+0, 3              # int(input()), int(input())
print(joseph_countdown_list(n, k))
print(joseph_countdown_list(n, k)[-1])
```

```python
### 2.2  для любого k    только последний выживший

def joseph_countdown(n, k):
    live_list = list(range(1, n+1))
    last = 0
    cnt = i = 0

    while live_list:
        cnt += 1
        if cnt % k == 0:            # жребий выпал
            last = live_list[i]
            live_list[i] = 0        # сохранить выбывшего и занулить его в списке

        i += 1
        if i == len(live_list):      # цикл закончился - обновить
            while 0 in live_list:
                live_list.remove(0)  # очистить следы выбывших
            i = 0                    # обнулить индикатор

    return last


n, k = 8+0, 2              # int(input()), int(input())
print(joseph_countdown(n, k))
```
