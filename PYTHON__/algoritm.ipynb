```python
# (( ?  алгоритм сортировки выбором   с двух сторон ))
a = [8,    5,  0, -1,  -67,  -4, 6,]

n = len(a)

for i in range(n//2):
    ind_mx = ind_mn = i
    for j in range(i+1, n - i):
        if a[j] > a[ind_mx]:
            ind_mx = j
        if a[j] < a[ind_mn]:
            ind_mn = j
    a[i], a[ind_mn] = a[ind_mn], a[i]
    if ind_mx == i:
        ind_mx = ind_mn
    a[n-i-1], a[ind_mx] = a[ind_mx], a[n-i-1]

print(a)
```
