```python
class Student:      # класс  (( конструктор ))
    name = 'Alex'
    age = 50
    def fun(self, x):
        return x*100

a = Student()      # экземпляр (объект)

print(Student.name, Student.age)
print(type(a))
print(type(Student))

print(a.fun(5))  
```
